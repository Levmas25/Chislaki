```python
# Задание 2
from math import log10
f = lambda x: 2*x - log10(x) - 2 # 2x - lg(x) - 2 = 0
e = 10**-3

g = lambda x: (log10(x) + 2) / 2 # x = (lg(x) + 2) / 2
```

```python
from utils import Utils
from matplotlib import pyplot as plt


def g_plot(end=2, label='f(x)=(lg(x)-2)/2'):
    Utils.create_plot(f=g, label=label, x_range=(10**-7, end))

def x_plot(end=2):
    plt.plot([10**-7, end], [10**-7, end], label='y=x', color='red')

g_plot()  
x_plot()
plt.legend()
plt.ylim((-1, 2))
plt.show()
```

```python
def simple_iteration(g, x_n, n):
    while abs(g(x_n) - x_n) > e:

        plt.plot(x_n, g(x_n), 'ro', label=f'x{n}={x_n:.5f}')
        plt.legend()
        n += 1
        x_n = g(x_n)
    plt.plot(x_n, g(x_n), 'ro', label=f'x{n}={x_n:.5f}')
    plt.legend()
```

```python
def newton_method(x_k, x_n, n):
    while abs(x_k(x_n) - x_n) > e:
        plt.plot(x_n, g(x_n), 'ro', label=f'x{n}={x_n:.5f}')
        plt.legend()
        x_n = x_k(x_n)
        n += 1
    plt.plot(x_n, g(x_n), 'ro', label=f'x{n}={x_n:.5f}')
    plt.legend()
```

```python
# Simple iteration
n = 0
x_n = 2

g_plot()
x_plot()

simple_iteration(g, x_n, n)

plt.legend()
plt.ylim((-1, 2))
plt.show()
```

```python
# Newton method
from math import log


f1 = lambda x: 2 - 1 / (x * log(10)) # f' = 2 - 1(x*ln(x))
x_k = lambda x: x - (f(x) / f1(x))
x_n = 0.5
n = 0

g_plot()
x_plot()

newton_method(x_k, x_n, n)

plt.ylim((-1, 2))
plt.show()
```

```python
# little changing in g so we don`t need to think about lg anymore
g = lambda x: 10**(2*x - 2) # g = 10^(2x - 2)
x_n = 10**-7
n = 0

g_plot(0.25, label="10^(2x-2)")
x_plot(0.25)

plt.ylim((-0.5, 0.5))
plt.title("Second dot")

simple_iteration(g, x_n, n)
```
