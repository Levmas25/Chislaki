```python
# Вариант 12 Задание 1
# Метод половинного деления
f = lambda x: x**3 - 2*x + 4 # = 0
e = 10**-3
```

```python
from utils import Utils
from matplotlib import pyplot as plt
import numpy as np

Utils.create_plot(f)
plt.show()
```

```python
# Возьмём а=-4 и b=1, так как очевидно, что корень находится между этих значений
a = -5
b = 2

Utils.create_plot(f)

i = 0
while abs(b - a) > e:
    
    c = (a + b) / 2
    plt.plot(c, f(c), 'ro', label=f'x{i}={c:.3f}')
    plt.legend()
    
    if f(c)*f(a) < 0:
        b = c
    else:
        a = c
    
    i += 1

plt.show()
```
